### Import Data

```python
path = os.getcwd()
path = path.replace('data_preparation_and_EDA', 'resources')
files = os.listdir(path)
for file in files:
    if len(file.split('.csv'))>1:
        csv_path = path+'/'+file
data = pd.read_csv(csv_path)
```
