# Data Preparation

### Load libraries

```python
import pandas as pd
import numpy as np
import os
```

### Import Data

```python
path = os.getcwd()
path = path.replace('data_analysis', 'resources')
files = os.listdir(path)
for file in files:
    if len(file.split('.csv'))>1:
        csv_path = path+'/'+file
data = pd.read_csv(csv_path)
```

```python
data['SeniorCitizen']=data['SeniorCitizen'].map({0:'No', 1:'Yes'})
data['TotalCharges']=pd.to_numeric(data.TotalCharges, errors='coerce')
```

```python
data.isnull().sum()
```

```python
data[data.TotalCharges.isnull()]
```

We can see that the Tenure is 0 for the observations were Total Charges is missing. Also, all of these seem to come under 'Yearly' contract. We can remove these from the data as they constitute a small fraction of the data, and they are all non-churners.

```python
data.drop(data[data.TotalCharges.isna()].index, axis=0, inplace=True)
```

We can check if there are other rows with 0 tenure

#### Remove unnecessary features

Customer ID just assigns a unique identifier to each customer. In the absence of other dataframes, this attribute is not required.

```python
data.drop('customerID', axis=1, inplace=True)
```

```python
for col in data.select_dtypes('object').columns:
    print(data[col].value_counts())
```

From the above, there seems to enough data in all categores to avoid deleting any of them. 

It can be noted that the presence or absence of `PhoneService` is already captured by `MultipleLines` so we can remove after EDA.

In case of `InternetService`, although the other column provide information on the absence of internet service, the feature tells us where the connection type is `DSL` or `Fiber optic`
